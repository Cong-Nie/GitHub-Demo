```python
import numpy as np
import pandas as pd
```

```python
def KS(n,C,wt,val):
    '''Build two matrices to record
    '''
    arr = np.zeros((n+1,C+1))
    item = np.zeros((n+1,C+1))
    for i in range(1,n+1):
        for j in range(1,C+1):
            '''If the weight that can hold is less than the item's weight, then keep last time
            '''
            if wt[i-1] <= j:
                arr[i][j] = max(val[i-1] + arr[i-1][j-wt[i-1]], arr[i-1][j])
                if val[i-1] + arr[i-1][j-wt[i-1]] > arr[i-1][j]:
                    '''This item is included
                    '''
                    item[i][j] = 1
            else:
                arr[i][j] = arr[i-1][j]         
    size = C
    result = []
    '''Record the chosen items in the (0,1) format
    '''
    for i in range(n,0,-1):
        if item[i][size] == 1:
            result.append(i)
            size -= wt[i-1]         
    result_ = [0] * n
    for i in result:
        result_[i-1] = 1
    '''Return the chosen item list as well as the optimal value
    '''
    return result_, arr[n][C]
```
